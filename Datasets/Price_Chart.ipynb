```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import scipy.stats as stats
```

```python
countries_data = pd.read_csv("country_by_variety.csv")
countries_data
```

```python
reduced_countries_data = countries_data[["country","variety","price"]]
reduced_countries_data
```

```python
#groupbys for dataset
country_groupby = reduced_countries_data.groupby(["country"]).mean()
```

```python
price_chart = country_groupby.plot(kind='bar', cmap='PuRd_r', legend=False)
price_chart.set_xlabel("country")
price_chart.set_ylabel("price")
plt.show()



```
